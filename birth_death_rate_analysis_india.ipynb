```python
plt.figure(figsize=(8,5))

sns.lineplot(data=data, x="Year", y="Birth_Rate", hue="State", marker="o")

plt.title("Birth Rate Trend: Uttar Pradesh vs Sikkim")
plt.xlabel("Year")
plt.ylabel("Birth Rate per 1000 population")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.lineplot(data=data, x="Year", y="Death_Rate", hue="State", marker="o")

plt.title("Death Rate Trend: Uttar Pradesh vs Sikkim")
plt.xlabel("Year")
plt.ylabel("Death Rate per 1000 population")

plt.show()
```

```python
data["Growth"] = data["Birth_Rate"] - data["Death_Rate"]

plt.figure(figsize=(8,5))

sns.barplot(data=data, x="Year", y="Growth", hue="State")

plt.title("Natural Population Growth Comparison")
plt.ylabel("Growth Rate")

plt.show()
```

```python
avg_birth = data.groupby("State")["Birth_Rate"].mean()

avg_birth.plot(kind="bar")

plt.title("Average Birth Rate by State")
plt.ylabel("Birth Rate")

plt.show()
```

```python
sns.boxplot(data=data, x="State", y="Birth_Rate")

plt.title("Birth Rate Distribution by State")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.lineplot(data=data, x="Year", y="Birth_Rate", hue="State", marker="o")
sns.lineplot(data=data, x="Year", y="Death_Rate", hue="State", marker="s", linestyle="--")

plt.title("Birth Rate vs Death Rate Trends")
plt.ylabel("Rate per 1000 population")

plt.show()
```

```python
corr = data[["Birth_Rate","Death_Rate"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Between Birth Rate and Death Rate")

plt.show()
```
